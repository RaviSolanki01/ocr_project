
Image sharpening
=================

This example shows how to sharpen an image in noiseless situation by
applying the filter inverse to the blur.



```python
import scipy
from scipy import ndimage
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

path ="./path/to/"
#path ="./image_sample/"
file = "b.jpg"

f = scipy.misc.imread(name = path + file,flatten = True).astype(float)
blurred_f = ndimage.gaussian_filter(f, 3)

filter_blurred_f = ndimage.gaussian_filter(blurred_f, 1)

alpha = 30
sharpened = blurred_f + alpha * (blurred_f - filter_blurred_f)


plt.figure(figsize=(12, 4))

plt.subplot(131)
plt.imshow(f, cmap=plt.cm.gray)
plt.axis('off')
plt.subplot(132)
plt.imshow(blurred_f, cmap=plt.cm.gray)
plt.axis('off')
plt.subplot(133)
plt.imshow(sharpened, cmap=plt.cm.gray)
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import pytesseract
text = pytesseract.image_to_string(sharpened, lang = 'eng')
print("**"+text)
```
