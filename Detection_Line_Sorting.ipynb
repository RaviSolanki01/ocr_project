```python
from PIL import Image
import pytesseract
import numpy

path = '/home/ravi/programs/OCR/image_sample/1.jpg'

im = Image.open(path)
```

```python
text = pytesseract.image_to_string(im, lang = 'eng')

print(text)
```

```python
text = pytesseract.image_to_boxes(im, lang = 'eng')
myList = []
aList = []
wrd = []
idx = ''
```

```python
for dtl in text:
    if dtl != ' ' and dtl != '\n':
        idx += dtl
    elif dtl == ' ' or dtl =='\n':
        wrd.append(idx)
        idx = ''
        if dtl == '\n':
            aList.append(wrd)
            wrd = []
print(aList)
```
