```python
import pillowfight
import cv2
import numpy as np
from matplotlib import pyplot as plt
import imutils
from PIL import Image

%matplotlib inline

file_path = '../path/to/b.jpg'
input_img = Image.open("../path/to/297622x.jpg?h=184494cd")
input_img1 = Image.open("../path/to/a.jpg")
input_img2 = Image.open("../image_sample/1.jpg")
```

```python

output_img_ace = pillowfight.ace(input_img)

import pytesseract
text = pytesseract.image_to_string(output_img_ace, lang = 'eng')
print("**"+text)
output_img_ace.save("new.jpg")



plt.subplot(1, 2, 1)
plt.imshow(input_img)
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(output_img_ace)
plt.title('Filtered Image')

plt.show()
```

```python
#ACE (Automatic Color Equalization ; Parallelized implementation)
output_img_ace = pillowfight.ace(input_img,
    slope=10,
    limit=1000,
    samples=100,
    seed=None)

plt.subplot(1, 2, 1)
plt.imshow(input_img)
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(output_img_ace)
plt.title('Filtered Image')

plt.show()
```

```python

output_img_canny = pillowfight.canny(input_img)

plt.subplot(1, 2, 1)
plt.imshow(input_img)
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(output_img_canny)
plt.title('Filtered Image')

plt.show()
```

```python
(nb_diff, output_img_compare) = pillowfight.compare(input_img, output_img_canny, tolerance=10)

plt.subplot(1, 3, 1)
plt.imshow(input_img)
plt.title('first Image')

plt.subplot(1, 3, 2)
plt.imshow(output_img_canny)
plt.title('second Image')

plt.subplot(1, 3, 3)
plt.imshow(output_img_compare)
plt.title('compared Image')

plt.show()
```

```python
output_img_gaussian = pillowfight.gaussian(input_img, sigma=2.0, nb_stddev=5)

plt.subplot(1, 2, 1)
plt.imshow(input_img)
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(output_img_gaussian)
plt.title('Filtered Image')

plt.show()
```

```python
output_img_sobel = pillowfight.sobel(input_img2).astype(float)

output_img_sobel = output_img_sobel - input_img2.astype(float)

plt.subplot(1, 2, 1)
plt.imshow(input_img2)
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(output_img_sobel)
plt.title('Filtered Image')


plt.show()
```

```python
output_image_swt = pillowfight.swt(input_img,
                                   output_type=pillowfight.SWT_OUTPUT_ORIGINAL_BOXES)

plt.subplot(1, 2, 1)
plt.imshow(input_img)
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(output_image_swt)
plt.title('Filtered Image')

plt.show()
```

```python
output_image_unpaper_blackfilter = pillowfight.unpaper_blackfilter(input_img)
output_image_unpaper_blurfilter = pillowfight.unpaper_blurfilter(input_img)
output_image_unpaper_border = pillowfight.unpaper_border(input_img)


plt.subplot(1, 2, 1)
plt.imshow(input_img)
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(output_image_unpaper_blackfilter)
plt.title('unpaper_blackfilter')

plt.show()

plt.subplot(1, 2, 1)
plt.imshow(output_image_unpaper_blurfilter)
plt.title('unpaper_blurfilter')

plt.subplot(1, 2, 2)
plt.imshow(output_image_unpaper_border)
plt.title('unpaper_blurfilter')

plt.show()
```

```python
output_imag_unpaper_grayfilter = pillowfight.unpaper_grayfilter(input_img)


plt.subplot(1, 2, 1)
plt.imshow(input_img)
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(output_imag_unpaper_grayfilter)
plt.title('Filtered Image')

plt.show()
```

```python
output_image_unpaper_masks = pillowfight.unpaper_masks(input_img)

plt.subplot(1, 2, 1)
plt.imshow(input_img)
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(output_image_unpaper_masks)
plt.title('unpaper_masks')

plt.show()
```
